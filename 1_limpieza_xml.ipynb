```python
import re
import os
import xml.etree.ElementTree as ET

tree = ET.parse('data_xml.xml')
root = tree.getroot()


xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
## Cambiamos el value de la key gender por el texto correspondiente en un solo diccionario para hacer la prueba

prueba = {'level_0': '1', 'index': '1', 'time': '784', 'age': '50-54', 'gender': '0'}

if prueba["gender"] == '0':
    prueba.update({"gender" : "Man"})
elif prueba["gender"] == '1':
    prueba.update({"gender" : "Woman"})
elif prueba["gender"] == '2':
    prueba.update({"gender" : "Nonbinary"})
elif prueba["gender"] == '3':
    prueba.update({"gender" : "Prefer not to say"})
else:
    prueba.update({"gender" : "Prefer to self-describe"})
     
print(prueba)
```

```python
## Hacemos un bucle for para cambiar la variable en todos los diccionarios de la lista xml_list

for i in xml_list:
      
   if i["gender"] == '0':
      i["gender"] = "Man"
   elif i["gender"] == '1':
      i["gender"] = "Woman"
   elif i["gender"] == '2':
      i["gender"] = "Non binary"
   elif i["gender"] == '3':
      i["gender"] = "Prefer not to say"
   else:
      i["gender"] = "Prefer to self-describe"

    
    
```

```python
## Probamos a modificar en el diccionario prueba los valores de las keys 'level_0' e 'index' por 'index_sql' e 'index_xml' respectivamente 

# primero hacemos una copia del diccionario

prueba2 = prueba.copy()
print(prueba2)

prueba2["index_sql"] = prueba2.pop("level_0")
print(prueba2)

prueba2["index_xml"] = prueba2.pop("index")
print(prueba2)
```
