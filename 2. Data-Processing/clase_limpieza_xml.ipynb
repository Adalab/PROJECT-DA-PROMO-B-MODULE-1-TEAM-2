```python
import re
import os
import xml.etree.ElementTree as ET
import mysql.connector
import pandas as pd

tree = ET.parse('data_xml.xml')
root = tree.getroot()


xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
# Definimos una clase que nos limpie archivos xml

class Limpieza_xml:
    
    def __init__ (self, xml):
        self.xml = xml
    
    def limpieza_gender(self):
    
        for i in self.xml:
              
            if i["gender"] == '0':
                i["gender"] = "Man"
            elif i["gender"] == '1':
                i["gender"] = "Woman"
            elif i["gender"] == '2':
                i["gender"] = "Non binary"
            elif i["gender"] == '3':
                i["gender"] = "Prefer not to say"
            else:
                i["gender"] = "Prefer to self-describe"
                
        return self.xml
    
    def cambiar_columnas (self, antigua_key, nueva_key):
        for i in self.xml:
            i[nueva_key] = i.pop(antigua_key)
    
            
        return self.xml
```

```python
# Creamos una instancia con la informaci√≥n de nuestro cliente

limpieza_adalab = Limpieza_xml(xml_list)
```

```python
# Cambiamos la variable g√©nero en los datos de nuestro cliente

limpieza_adalab.limpieza_gender()
```

```python
# Cambiamos el nombre de la columna index

limpieza_adalab.cambiar_columnas("index", "index_xml")
```

```python
# Cambiamos el nombre de la columna level_0

limpieza_adalab.cambiar_columnas("level_0", "index_sql")
```

```python
#Almacenamos el archivo modificado en una nueva variable

xml_limpio = limpieza_adalab.xml
```

```python
# Transformamos la lista de diccionarios a lista de tuplas desde un xml

def pasar_lista_tuplas_xml(archivo):

    lista_tuplas = []

    for v in archivo:
        x = tuple(v.values())
        lista_tuplas.append(x)
    return lista_tuplas    
    
    
```

```python
val = pasar_lista_tuplas_xml(xml_limpio)
```

```python
# Creamos una funci√≥n para insertar datos desde Python

def insertar_datos(sql, val):    
     
    cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                                host='127.0.0.1', database='project1')

    mycursor = cnx.cursor()
    
    try: 
        mycursor.executemany(sql, val)
        cnx.commit()
        print(mycursor.rowcount, "registro insertado.")

    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

```python
insertar_datos(sql, val)
```

```python
class Lectura:
    
    def __init__(self, password):
        self.password = password
      
        
    def lectura_df (self, query, lista_columnas_df):
        self.query = query
        self.lista_columnas_df = lista_columnas_df
        
        cnx = mysql.connector.connect(user='root', password=self.password,
                              host='127.0.0.1', database='project1')

        mycursor = cnx.cursor()
        mycursor.execute(self.query)
        myresult = mycursor.fetchall()
        df = pd.DataFrame(myresult, columns = self.lista_columnas_df)
        
        return print(df)
```
